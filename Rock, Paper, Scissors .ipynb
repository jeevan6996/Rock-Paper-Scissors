```python
from getpass import getpass
from random import randint
```

```python
def choose(player):
    while True:
        try:
            choice = int(getpass(f"{player} choose your option \n1. Rock \n2. Paper \n3. Scissors \n:"))
            if choice not in [1, 2, 3]:
                print("Please enter either 1, 2 or 3.")
                continue
            break
        except:
            print("Please enter a valid choice.")
    return choice
```

```python
choice_val = { 1 : "Rock", 2 : "Paper", 3 : "Scissors"}

def check_winner(game_type, player1_choice, player2_choice):
    if player1_choice == player2_choice:
        print(f"Both chose {choice_val[player1_choice]}")
        print("It's a DRAW!")
        return
    elif player1_choice == 1 and player2_choice == 2:
        winner = 2
    elif player1_choice == 1 and player2_choice == 3:
        winner = 1
    elif player1_choice == 2 and player2_choice == 1:
        winner = 1
    elif player1_choice == 2 and player2_choice == 3:
        winner = 2
    elif player1_choice == 3 and player2_choice == 1:
        winner = 2
    elif player1_choice == 3 and player2_choice == 2:
        winner = 1
    if game_type == 1:
        print(f"CPU chose {choice_val[player1_choice]}")
        print(f"Player 1 chose {choice_val[player2_choice]}")
        if winner == 1:
            print("CPU wins!")
        else:
            print("Player 1 wins!")
    else:
        print(f"Player 1 chose {choice_val[player1_choice]}")
        print(f"Player 2 chose {choice_val[player2_choice]}")
        print(f"Player {winner} wins!")
```

```python
play_game = 1

while play_game == 1:
    while True:
        print("\nWelcome to Rock, Paper, Scissors Game!")
        print("1. Play against Computer \n2. Play against Human")
        try:
            game_type = int(input("Enter your choice : "))
            if game_type not in [1, 2]:
                print("Please enter either 1 or 2.")
                continue
            break
        except:
            print("Please enter a valid choice.")
        
    if game_type == 1:
        cpu_choice = randint(1,3)
        player1_choice = choose('Player 1')
        check_winner(game_type, cpu_choice, player1_choice)
    else:
        player1_choice = choose('Player 1')
        player2_choice = choose('Player 2')
        check_winner(game_type, player1_choice, player2_choice)
    
    while play_game not in ['y', 'n']:
        play_game = input("Play another game? (Y/N) : ").lower()
        if play_game == 'y':
            play_game = 1
            break
        elif play_game == 'n':
            play_game = 0
            break
```
